```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px

#seaborn package
Xpoints=np.array([0,20,120,200])
Ypoints=np.array([10,80,100,120,150,180,250])

# Fix: Create a DataFrame 'df' from Ypoints to resolve the NameError
df = pd.DataFrame({'column2': Ypoints})
df['column1'] = df.index.values # Using index as 'column1' for x-axis

print(sns.lineplot(data=df, x="column1", y="column2"))
print(plt.show())
```

```python
print(sns.scatterplot(data=Ypoints))
print(plt.show())
```

```python
print(sns.barplot(data=Ypoints))
print(plt.show())
```

```python
#3-D plot
X=np.array([1,2,6,8,12,14,20])
Y=np.array([3,8,1,10,12,16,18])
# For a 3D scatter plot, a Z array is required.
# Assuming a simple sequence for Z values as an example. Adjust Z based on your actual data.
Z=np.arange(len(X))

Fig=plt.figure(figsize=(12,8))
Ax=Fig.add_subplot(111, projection='3d') # Corrected 'prohection' to 'projection' and using Fig.add_subplot
Plt_3d=Ax.scatter3D(X,Y,Z, c=Z, cmap='viridis') # Corrected 'ax' to 'Ax', 'x,y' to 'X,Y', added Z, and 'c' for colorbar
Fig.colorbar(Plt_3d, ax=Ax, shrink=0.5, aspect=5) # Corrected colorbar creation for 3D plot
plt.show() # Corrected 'Plt' to 'plt'

#plotly package
# Note: The X array is redefined here, which might not be intended if it's meant to be related to the 3D plot's X.
X=np.array([10,80,100,120,150,180,250])
```
