```python
import pandas as pd
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],
                index=['one','two','three'],
                columns=['a','b','c'])
print(df)
```

```python
#Taking the input from DataFrame anf storing in the Exel file
print("Taking the input from data frame and storing in the exel file")
df.to_excel('pandas_to_excel.xlsx',sheet_name='new_sheet_name')
```

```python
#second dataframe input to another Exel file
d=pd.DataFrame([[110,210,320],[12,220,320],[310,320,330]],index=['four','five','six'],columns=['a','b','c'])
d.to_excel('pandas_to_excell.xlsx',sheet_name='new_sheet_name')
```

```python
#merging two Excel files input into third file
x=pd.read_excel('pandas_to_excel.xlsx')
y=pd.read_excel('pandas_to_excell.xlsx')
z=pd.concat([x,y])
z.to_excel('pandas_to_excel3.xlsx')
```

```python
#sorting the column values
df=z.sort_values(["a"])
print(df)
df.to_excel('pandas_to_excel4.xlsx')
```
