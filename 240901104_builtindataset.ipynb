```
from sklearn import datasets

iris = datasets.load_iris()

print("type:")
print(type(iris))

print("\nkeys:")
print(iris.keys())

print("\ntype of data and target:")
print(type(iris.data), type(iris.target))

print("\ndata shape:")
print(iris.data.shape)
```
