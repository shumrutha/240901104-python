```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original data:\n")
df=pd.read_excel("studentsmark.xlsx")
print(df)
```

```python

f=df['Student Name'].str.strip()

f.to_excel("studentsmark.xlsx")

print("\nReplace value:")
k=df.ffill()
print(k)
print("\nUpdated list\n")
k.to_excel("studentsmark.xlsx")
```

```python

print("\nDrop one particular column and its values:\n")

print(k)
print("\nUpdated list..\n")
k.to_excel(r"D:\dcap.xlsx")
```

```python
print("\nDrop NaN rows:\n")
df=pd.read_excel("studentsmark.xlsx")
print(df)
x=df.dropna()
print(x)

x.to_excel("studentsmark.xlsx")
```

```python

print("\nReplace values:\n")
n=pd.read_excel("studentsmark.xlsx")
print("\nOriginal data:\n")
print(n)
y=n.replace({89:88})
print("\nReplaced dataset:\n")
print(y)

y.to_excel("studentsmark.xlsx")
```

```python
df=pd.read_excel("studentsmark.xlsx")
print(df)
print("\nExtract using isin() function:\n")
new=df['Eng'].isin([88])
print(df[new])
```
