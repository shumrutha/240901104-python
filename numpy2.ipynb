```
import numpy as np
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
print("Original array:\n",arr)
```

```
#returns every other rows
print("\n Every other rows:\n",arr[0:3:2])
```

```
#returns every other element from the entire array
arr=np.array([1,2,3,4,5,6,7])
print("\n original array:",arr)
print("\n original every other element in the array:arr[::2]:",arr[::2])
```

```
#slicing array
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
temp=arr[:2,:3]
print("\nArray with first 2 rows and 3 columns:\n",temp)
```

```
#integer array indexing
temp1=arr[[0,1,2,3],[3,2,1,0]]
print("\nElements at indices(0,3),(1,2),(2,1),""(3,0):\n",temp1)
```

```
#boolean array indexing
cond=arr>2
#cond is a boolean array
temp2=arr[cond]
print("\n elements greater than 2:\n",temp2)
```
